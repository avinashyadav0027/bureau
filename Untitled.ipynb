```python
 def clean_text_column(column):
        return column.str.strip().str.replace(' ', '_').str.lower()

def basic_cleaning(df,temp):
    
    df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))  

    object_cols = df.select_dtypes(include='object').columns
   
    for col in object_cols:
        df[col] = clean_text_column(df[col])

    if(temp):
        threshold = int(0.3 * len(df))
        print(threshold)
        df = df.dropna(thresh=threshold, axis=1)
    
    return df

df = basic_cleaning(df,True)
df.head()
```

```python
chi_sq_needed = []
for cols in df.drop(['application_status'], axis = 1).columns:
    mi = mutual_info_regression(df[cols].values.reshape(-1, 1), df['application_status'])
    if(mi < 0.01):
        chi_sq_needed.append(cols)
    print(f"Mutual Information for {cols}: {mi[0]:.4f}")
```

```python
from sklearn.feature_selection import chi2

remove_cols = []

for cols in chi_sq_needed:
    if (df[cols] >= 0).all():
        chi2_values, p_values = chi2(df[[cols]], df['application_status'])
        print(cols, ": Chi-Square values:", chi2_values, "  :P-values:", p_values )
        if(p_values >= 0.05):
            remove_cols.append(cols)
    else:
        remove_cols.append(cols)
        print(cols, ": negative value found")
```

```python
def transform_dataframe(dft):
    dft = basic_cleaning(dft,False)
    
    for col in dft.columns:
        if(col.startswith('phone_social_premium')):
            dft[col] = dft[col].astype(bool)
            dft[col].fillna(dft[col].mode(), inplace = True)

    dft = dft.drop(['application_login_date','first_name','middle_name','last_name','aadhar_verified','mobile_verification',
                    'phone_social_premium.rummycircle','phone_social_premium.my11','phone_social_premium.jiomart','phone_social_premium.a23games',
                    'phone_social_premium.whatsapp','phone_social_premium.iswabusiness','personal_email_address','mobile', 'dealer_name',
                     'phone_social_premium.yatra','dob', 'pan_name', 'name','upi_name'],axis = 1)
    
    dft['cibil_score'] = dft['cibil_score'].replace('-', np.nan).replace('no_response',np.nan).astype(float)
    
    dft['age'] = dft['age'].replace(0,np.nan)
    dft['phone_digitalage'] = dft['phone_digitalage'].replace(-1,np.nan)
    dft['phone_namematchscore'] = dft['phone_namematchscore'].replace(-1,np.nan)

    numerical_cols = ['applied_amount','age','phone_digitalage','phone_namematchscore', 'total_asset_cost']

    for col in numerical_cols:
        median_val = df[col].median()
        if(col != 'cibil_score'): dft[col].fillna(median_val, inplace=True)

    # dft = dft.drop(['aadhar_verified', 'mobile_verification'], axis = 1)
    categorical_cols = dft.select_dtypes(include=['object','bool']).columns
    dft['address_type'] = dft['address_type'].replace(address_mapping)

    categorical_cols = dft.select_dtypes(include=['object','bool']).columns

    for col in categorical_cols:
        le = label_encoders[col]
        
        # Replace unseen categories with 'unknown' before transforming
        dft[col] = dft[col].apply(lambda x: le.transform([x])[0] if x in le.classes_ else df[col].mode()[0])

    dft['age_group'] = pd.cut(dft['age'], bins=[0, 25, 35, 45, 100], labels=[0, 1, 2, 3]).astype(int)
    dft['employer_size'] = dft.groupby('employer_name')['employer_name'].transform('count')
    dft['asset_financed_ratio'] = dft['total_asset_cost'] / dft['applied_amount']

    ecommerce_sites = [
        'phone_social_premium.amazon',
        'phone_social_premium.byjus',
        'phone_social_premium.flipkart',
        'phone_social_premium.housing',
        'phone_social_premium.indiamart'
    ]
    
    marital_sites = [
        'phone_social_premium.jeevansaathi',
        'phone_social_premium.shaadi'
    ]
    
    social_media = [
        'phone_social_premium.instagram',
        'phone_social_premium.skype',
        'phone_social_premium.toi'
    ]
    
    payment_sites = [
        'phone_social_premium.paytm'
    ]
    
    other_sites = [
        'phone_social_premium.microsoft',
        'phone_social_premium.zoho'
    ]

    dft['ecommerce_sum'] = dft[ecommerce_sites].sum(axis=1)
    dft['marital_sum'] = dft[marital_sites].sum(axis=1)
    dft['social_media_sum'] = dft[social_media].sum(axis=1)
    dft['payment_sum'] = dft[payment_sites].sum(axis=1)
    dft['other_sum'] = dft[other_sites].sum(axis=1)
    
    dft = dft.drop(columns=ecommerce_sites + marital_sites + social_media + payment_sites + other_sites)

    dft['marital_status'] = dft.apply(predict_marital_status, axis=1)
    dft['address_type'] = dft.apply(predict_address_type, axis = 1)
    dft['employer_type'] = df.apply(predict_employer_type, axis=1)
    dft['cibil_score'] = df.apply(predict_cibil_score, axis = 1)

    return dft
```
